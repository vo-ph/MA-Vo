```python
import pandas as pd
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# constants  #
#------------#
HDF5_dir = "E:\\toResample\\"
resampled_dir = r'E:\resampled_phasevise'
dates = os.listdir(HDF5_dir)
phases = ["1","2","3"]
```

```python
for date in dates:
    print(date)
    with h5py.File(HDF5_dir + date + "\\" + "harmonics_phase_" + "1" + "_" + date + ".h5", 'r') as f:
        ts = pd.to_datetime(f["tenperiodtable"]["timestamp"], unit="s").tz_localize("UTC").tz_convert("Europe/Berlin")
        hu1df = pd.DataFrame(f["tenperiodtable"]["harmonics_u"][:,38:], ts)
        hu1df.index = ts
        hu1df_rs_mean = hu1df.resample("10T").mean()
        hu1df_rs_max = hu1df.resample("10T").max()
        
        hu1df_rs_mean.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_mean_p1.p")
        hu1df_rs_max.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_max_p1.p")
        
        del hu1df
        hi1df = pd.DataFrame(f["tenperiodtable"]["harmonics_i"][:,38:], ts)
        hi1df.index = ts
        hi1df_rs_mean = hi1df.resample("10T").mean()
        hi1df_rs_max = hi1df.resample("10T").max()
        
        hi1df_rs_mean.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_mean_p1.p")
        hi1df_rs_max.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_max_p1.p")
        
        del hi1df
    with h5py.File(HDF5_dir + date + "\\" + "harmonics_phase_" + "2" + "_" + date + ".h5", 'r') as f:
        hu2df = pd.DataFrame(f["tenperiodtable"]["harmonics_u"][:,38:], ts)
        hu2df.index = ts
        hu2df_rs_mean = hu2df.resample("10T").mean()
        hu2df_rs_max = hu2df.resample("10T").max()
        
        hu2df_rs_mean.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_mean_p2.p")
        hu2df_rs_max.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_max_p2.p")
        
        del hu2df
        hi2df = pd.DataFrame(f["tenperiodtable"]["harmonics_i"][:,38:], ts)
        hi2df.index = ts
        hi2df_rs_mean = hi2df.resample("10T").mean()
        hi2df_rs_max = hi2df.resample("10T").max()
        
        hi2df_rs_mean.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_mean_p2.p")
        hi2df_rs_max.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_max_p2.p")
        
        del hi2df
    with h5py.File(HDF5_dir + date + "\\" + "harmonics_phase_" + "3" + "_" + date + ".h5", 'r') as f:
        hu3df = pd.DataFrame(f["tenperiodtable"]["harmonics_u"][:,38:], ts)
        hu3df.index = ts
        hu3df_rs_mean = hu3df.resample("10T").mean()
        hu3df_rs_max = hu3df.resample("10T").max()
        
        hu3df_rs_mean.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_mean_p3.p")
        hu3df_rs_max.to_pickle(resampled_dir + "\\" + "u" + "\\" + date + "_harmonics10min_u_max_p3.p")
        
        del hu3df
        hi3df = pd.DataFrame(f["tenperiodtable"]["harmonics_i"][:,38:], ts)
        hi3df.index = ts
        hi3df_rs_mean = hi3df.resample("10T").mean()
        hi3df_rs_max = hi3df.resample("10T").max()
        
        hi3df_rs_mean.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_mean_p3.p")
        hi3df_rs_max.to_pickle(resampled_dir + "\\" + "i" + "\\" + date + "_harmonics10min_i_max_p3.p")
        
        del hi3df
  
    #hu_concat_mean = pd.concat((hu1df_rs_mean, hu2df_rs_mean, hu3df_rs_mean))
    #hu_all_mean = hu_concat_mean.groupby(hu_concat_mean.index).mean()
    #hu_concat_max = pd.concat((hu1df_rs_max, hu2df_rs_max, hu3df_rs_max))
    #hu_all_max = hu_concat_max.groupby(hu_concat_max.index).max()
   
    #hi_concat_mean = pd.concat((hi1df_rs_mean, hi2df_rs_mean, hi3df_rs_mean))
    #hi_all_mean = hi_concat_mean.groupby(hi_concat_mean.index).mean()
    #hi_concat_max = pd.concat((hi1df_rs_max, hi2df_rs_max, hi3df_rs_max))
    #hi_all_max = hi_concat_max.groupby(hi_concat_max.index).max()
    
```
