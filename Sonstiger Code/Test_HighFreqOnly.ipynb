```python
import pandas as pd
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
with h5py.File("C:\MessdatenStoeckach\\2017-03-02\harmonics_phase_1_2017-03-02.h5", "r") as f:
    dset = np.array(f["tenperiodtable"])
    phase1harmonics = pd.DataFrame(dset['harmonics_i'])
phase1_2kHz = phase1harmonics.iloc[:,40:]
```

```python
model = PCA(2)
```

```python
phase1_trans = model.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(phase1_2kHz)
scaler.transform(phase1_2kHz)
phase1_trans = model.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
with h5py.File("C:\MessdatenStoeckach\\2017-03-02\harmonics_phase_1_2017-03-02.h5", "r") as f:
    dset = np.array(f["tenperiodtable"])
    phase1harmonics = pd.DataFrame(dset['harmonics_u'])
phase1_2kHz = phase1harmonics.iloc[:,40:]
phase1_trans = model.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
scaler = StandardScaler()
scaler.fit(phase1_2kHz)
scaler.transform(phase1_2kHz)
phase1_trans = model.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
from sklearn.decomposition import FactorAnalysis
```

```python
transformer = FactorAnalysis(n_components=2)
```

```python
phase1_trans = transformer.fit_transform(phase1_2kHz)
```

```python
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
from sklearn.decomposition import FastICA
```

```python
fastICA = FastICA(n_components=2)
```

```python
phase1_trans = fastICA.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```

```python
from sklearn.random_projection import GaussianRandomProjection
```

```python
transformer = GaussianRandomProjection(n_components=2)
```

```python
phase1_trans = transformer.fit_transform(phase1_2kHz)
plt.scatter(phase1_trans[:,0], phase1_trans[:,1])
plt.show()
```
